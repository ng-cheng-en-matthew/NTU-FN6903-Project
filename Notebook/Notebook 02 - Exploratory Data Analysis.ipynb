# EDA Graphs

```python
# make a copy of train data for visualisation
X_train_cp = X_train.copy()
X_train_cp['ONSET_DATE'] = pd.to_datetime(X_train_cp['ONSET_DATE'])
```

```python
MANU_LABELS = ['PFIZER', 'MODERNA', 'JANSSEN', 'NOVAVAX', 'UNKNOWN']
MANU_COLS = [f'MANU_{m}' for m in MANU_LABELS[:-1]]

# reconstruct categorical variable from dummies
X_train_cp['MANU_UNKNOWN'] = 1-X_train_cp[MANU_COLS].sum(axis=1) # identify unknown manufacturers
X_train_cp.columns = X_train_cp.columns.str.replace('MANU_', '') # rename manufacturers
X_train_cp['Manufacturer'] = X_train_cp[MANU_LABELS].columns[np.where(X_train_cp[MANU_LABELS]!=0)[1]].str.capitalize()
```

### Rate of Adverse Events over Time by Vaccine Manufacturer

```python
event_count_by_manu_df = (
    X_train_cp
    .groupby(['ONSET_DATE', 'Manufacturer'], as_index=False)
    .agg({'NUMDAYS': 'count'})
    .rename(columns={'NUMDAYS': 'No. of Adverse Events'})
    .sort_values('ONSET_DATE')
)
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(
    data=event_count_by_manu_df,
    x='ONSET_DATE',
    y='No. of Adverse Events',
    hue='Manufacturer',
    hue_order=['Pfizer', 'Moderna', 'Janssen', 'Novavax', 'Unknown'],
    linewidth=0.7
)
plt.xlabel('Reported Onset Date of Adverse Event')
plt.title('No. of Reported Adverse Events per Day over Time (Train Period)', fontsize=15)
plt.savefig(os.path.join(IMG_DIR, 'count_by_manu.pdf'), bbox_inches='tight')
plt.show()
```

### Rate of Adverse Events over Time by Hospitalisation Status

```python
event_count_by_hosp_df = pd.concat([X_train_cp, y_train], axis=1)
event_count_by_hosp_df = (
    event_count_by_hosp_df
    .groupby(['ONSET_DATE', 'HOSP'], as_index=False)
    .agg({'NUMDAYS': 'count'})
    .rename(columns={'NUMDAYS': 'No. of Adverse Events'})
    .sort_values('ONSET_DATE')
)
event_count_by_hosp_df['Hospitalised'] = np.where(event_count_by_hosp_df['HOSP']==1, 'Yes', 'No')
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(
    data=event_count_by_hosp_df,
    x='ONSET_DATE',
    y='No. of Adverse Events',
    hue='Hospitalised',
    hue_order=['Yes', 'No'],
    palette=['red', 'blue'],
    linewidth=0.7
)
plt.xlabel('Reported Onset Date of Adverse Event')
plt.title('No. of Reported Adverse Events per Day over Time (Train Period)', fontsize=15)
plt.savefig(os.path.join(IMG_DIR, 'count_by_hosp.pdf'), bbox_inches='tight')
plt.show()
```
